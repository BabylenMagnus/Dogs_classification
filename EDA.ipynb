## Show data

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import random
```

```python
plt.figure(figsize=(20, 10))

count = []
breeds = []

for i, breed in enumerate(os.listdir(TRAIN_DATA), 1):
    candidates = os.listdir(os.path.join(TRAIN_DATA, breed))    
    img = random.choice(candidates)
    img = mpimg.imread(os.path.join(TRAIN_DATA, breed, img))

    breed = breed.replace('_', ' ').title()
    breeds.append(breed)
    count.append(len(candidates))
  
    ax = plt.subplot(2, 5, i)
    plt.imshow(img)
    plt.title(breed, fontdict={'fontsize': 18})
    plt.axis("off")
```

```python
plt.figure(figsize=(10, 10))

plt.pie(count, labels=breeds, autopct='%1.1f%%')
plt.show()
```
